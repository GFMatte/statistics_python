### Using Z test to validate "notas_filmes_1.rating with 'movielens_average_10_or_more_votes.mean()'

```python
from statsmodels.stats.weightstats import ztest
movielens_average_10_or_more_votes.mean()
```

```python
zconfint(notas_filme_1.rating)
```

```python
tstat, p_value = ztest(notas_filme_1.rating, value = movielens_average_10_or_more_votes.mean())
print(p_value)
```

```python
print ('H0 rejected') if p_value < 0.05 else print ('H0 accepted')
```

### Analyzing movie1 mean variation during sample len variation

### Analyzing p_value and mean variation with sample len

```python
np.random.seed(75241)
temp = notas_filme_1.sample(frac=1).rating # Random sample of all items in movielens_.......

def calcula_teste(i):
    media = temp[0:i].mean()
    stat, p_value = ztest(temp[0:i], value = movielens_average_10_or_more_votes.mean())
    return (i, media, p_value)
    
valores = np.array([calcula_teste(i) for i in range(2, len(temp))])
valores
# figure(figsize=(12, 8), dpi=80)
# plt.plot(valores[:1])
```

## Testing a series of data for a normal distribuition

```python
# using scipy
from scipy.stats import normaltest
stat, p = normaltest(notas1.rating)
print(f'Stat: {stat.round(4)}, p_value: {p.round(4)}')
```

```python
if p < 0.05:
    print('H0 Rejected, series does not come from a normal dist.')
```

### Using a non-parametric test (ranksums / wilcoxon)

```python
from scipy.stats import ranksums
_, p_value = ranksums(notas1.rating, notas593.rating)
p_value.round(5)
print('H0 Rejected, values in one sample are more likely to be larger than the values in the other sample') if p_value < 0.05 else ""
```
